```python
import pandas as pd
import numpy as np
from scipy import stats
file= pd.read_csv("C:/Users/al6436/Desktop/Vrinda Store Data/VrindaStore.csv")
file.sort_values(by='Category', ascending=True, inplace=True)
```

Null Hypothesis (H₀): There is no association between gender and the number of orders across the categories.


Alternative Hypothesis (H₁): There is an association between gender and the number of orders across the categories.

```python
from scipy.stats import chi2_contingency
import pandas as pd

data = {'Men': [7, 46, 176, 179, 295, 5157, 52, 3623],
        'Women': [227, 32, 88, 10362, 1094, 7289, 2149, 461]}
df = pd.DataFrame(data, index=file['Category'].unique())
print(df)
chi2, p, dof, expected = chi2_contingency(df)
print(f'Chi-square value: {chi2}, p-value: {p}')
if(p<0.05):
    print("Reject the Null Hypothesis")
if(p>0.05):
    print("Fail to Reject Null Hypotheesis")
```

```python
import matplotlib.pyplot as plt

df.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Orders per Category by Gender')
plt.ylabel('Number of Orders')
plt.xlabel('Categories')
plt.xticks(rotation=0)
plt.show()
```
