Question Of Uniform Distribution

Suppose you are designing a lottery system where 1000 tickets are sold, and each ticket has a unique number between 1 and 1000. You need to write a Python function to determine the number of prizes each ticket should win if every ticket is equally likely to win a certain number of prizes, following a uniform distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as stats

num_tickets = 1000
low_prizes = 1
high_prizes = 50

prizes = np.random.uniform(low_prizes, high_prizes, num_tickets)

plt.hist(prizes, bins=30, density=True)
plt.title('Uniform Distribution of Prizes')
plt.xlabel('Number of Prizes')
plt.ylabel('Frequency')
plt.show()

for i in range(10):
    print(f"Ticket {i + 1}: {prizes[i]:.2f} $ prizes")

```

Question of Normal Distribution

You are a teacher and want to generate the test scores of 30 students in a class. The test scores are normally distributed with a mean of 70 and a standard deviation of 10. Write a Python function to generate these test scores. Compare the results by plotting histograms.

```python
mean = 70
std_dev = 10
size = 30

scores = np.random.normal(mean, std_dev, size)
print(scores)

plt.hist(scores, bins=10, density=True, alpha=0.6, color='green' )
```

```python
mean = 70
std_dev = 10
size = 30
num_samples = 30 

def normal(mean, std_dev, size, num_samples):
    normal_distribution = []
    for i in range(size):
        sum_uniform = sum(random.uniform(0, 1) for i in range(num_samples))
        z = (sum_uniform - num_samples / 2) / (num_samples / 15)**0.5  
        normal_distribution.append(mean + z * std_dev)
    return normal_distribution

scores = normal(mean, std_dev, size, num_samples)

plt.hist(scores, bins=10, density=True, alpha=0.6, color='purple', label='CLT')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (std_dev * (2 * np.pi)**0.5)) * np.exp(-0.5 * ((x - mean) / std_dev)**2)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.title('Normal Distribution of Test Scores')
plt.xlabel('Test Score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

for i in range(10):
    print(f"Test Score {i + 1}: {scores[i]:.2f}")
```

Question of Exponential Distribution 

You are analyzing the waiting times for a bus at a bus stop. The time between arrivals of buses follows an exponential distribution with an average wait time of 10 minutes. You need to generate a dataset of 30 waiting times.

```python
scale = 10
size = 30

waiting_times = np.random.exponential(scale, size)
for i in range(10):
    print(f"Waiting Time {i + 1}: {waiting_times[i]:.2f} minutes")
plt.hist(waiting_times, bins=10, density=True, alpha=0.6, color='brown')
```

```python
def exponential(scale, size):
    exponential_distribution = []
    for _ in range(size):
        u = random.random()
        x = -scale * math.log(u)
        exponential_distribution.append(x)
    return exponential_distribution

waiting_times = exponential(scale, size)

plt.hist(waiting_times, bins=10, density=True, alpha=0.6, color='yellow')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / scale) * np.exp(-x / scale)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.title('Exponential Distribution of Waiting Times')
plt.xlabel('Waiting Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

for i in range(10): 
    print(f"Waiting Time {i + 1}: {waiting_times[i]:.2f} minutes ")
```

Question of Poisson Distribution

You are analyzing the number of emails received per hour at a support desk. The number of emails received follows a Poisson distribution with an average rate of 5 emails per hour. You need to generate a dataset of the number of emails received in 30 hours

```python
lambda_rate = 5
size = 30
email_counts = np.random.poisson(lambda_rate, size)
for i in range(10):
    print(f"Hour {i + 1}: {email_counts[i]} emails")

plt.hist(email_counts, bins=range(min(email_counts), max(email_counts) + 2), density=True, alpha=0.6, color='red')
```

```python
lambda_rate = 5
size = 30

def poisson(lambda_rate, size):
    poisson_distribution = []
    for i in range(size):
        L = math.exp(-lambda_rate)
        k = 0
        p = 1
        while p > L:
            k += 1
            p *= random.random()
        poisson_distribution.append(k - 1)
    return poisson_distribution

email_counts = poisson(lambda_rate, size)

plt.hist(email_counts, bins=range(min(email_counts), max(email_counts) + 2), density=True, alpha=0.6, color='green')

x = range(min(email_counts), max(email_counts) + 1)
p = [(lambda_rate ** k * math.exp(-lambda_rate)) / math.factorial(k) for k in x]
plt.plot(x, p, 'k', linewidth=2, label='PMF')

plt.title('Poisson Distribution of Email Counts:')
plt.xlabel('Number of Emails')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

for i in range(10):
    print(f"Hour {i + 1}: {email_counts[i]} emails")
```

Geometrical Distribution

Imagine you're playing a game where you roll a fair die, and you want to know how many times you need to roll the die before you get a 6. The probability of rolling a 6 (success) is 
1/6.

```python
import numpy as np
import matplotlib.pyplot as plt

p = 1/6  
size = 50 

geometric = np.random.geometric(p, size)

plt.hist(geometric, bins=range(1, max(geometric) + 2), density=True, alpha=0.6, color='blue', edgecolor='black', align='left')
plt.title('Geometric Distribution ')
plt.xlabel('Number of Trials to First Success')
plt.ylabel('Frequency')
plt.show()

print(f'Trials to First Success: {geometric}')
```

```python
import random
import matplotlib.pyplot as plt

def geometric(p, size):
    geometric_distribution = []
    for i in range(size):
        trials = 0
        while random.random() > p:
            trials += 1
        geometric_distribution.append(trials + 1)
    return geometric_distribution

geometric_data = geometric(p, size)

plt.hist(geometric_data, bins=range(1, max(geometric_data) + 2), density=True, alpha=0.6, color='green', edgecolor='black', align='left')
plt.title('Geometric Distribution ')
plt.xlabel('Number of Trials to First Success')
plt.ylabel('Frequency')
plt.show()

print(f'Trials to First Success: {geometric_data}')
```
